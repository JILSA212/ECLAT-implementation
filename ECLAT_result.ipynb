```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/ML_project/new_data (6).csv")
df
```

```python
items = list(df["Unnamed: 0"].copy())
items
```

```python
recommendation = {}

for i in items:
  temp = list(i.split(","))
  temp_value = temp[-1]
  temp_key = list(temp[:-1])
  recommendation[str(temp_key)] = temp_value

print(recommendation)
```

```python
input_item = ""
input_list = []

while(input_item.lower() != "quit"):
  recommend_list = []
  input_item = input()
  if(input_item.lower() == "quit"):
    print("Thank you")
    break
  input_list.append(input_item)
  try:
    for i in recommendation.keys():
      if(all(j in i for j in input_list)):
        recommend_list.append(recommendation[i])
    if(len(recommend_list)==0):
      print("No recommendation")
      input_list = []
    else:
      print("Recommendations:")
      for i in set(recommend_list):
        print(i)
  except(e):
    print("Exception occured",e)
```
