```python
df = df[["InvoiceNo", "Description"]]
df["InvoiceNo"] = df["InvoiceNo"].astype("str")
```

```python
itemset = {}
temp = []

#print(list(df.loc[df["Description"] == "SET OF 6 T-LIGHTS SNOWMEN" ,"InvoiceNo"]))
unique_items = df["Description"].unique()

for i in unique_items:
#for k in range(0,1):
  #temp = []
  #temp.append(df.loc[df["Description"] == i ,"InvoiceNo"])
  #print(temp)
  itemset[i]=list(df.loc[df["Description"] == i ,"InvoiceNo"])

print(itemset)
```

```python
count_items = {}

for i in itemset.keys():
  count_items[i] = len(itemset[i])

print(count_items)
print(len(count_items))
```

```python
threshold = 0.0005
min_sup = int(threshold * df.shape[0])
print(min_sup)

removing_ele = []

for i in count_items:
  if(count_items[i] < min_sup):
    removing_ele.append(i)
    
print(len(removing_ele))

for i in removing_ele:
  count_items.pop(i)
  itemset.pop(i)

single_items = count_items.copy()
print(len(count_items))
```

```python
temp_dict = {}

for i in itemset.keys():
  for j in itemset.keys():
    if(i!=j):
      temp_set = ()
      temp_set = set(itemset[i]).intersection(set(itemset[j]))
      if(len(temp_set) > min_sup):
        count_items[str(i) + "," + str(j)] = len(temp_set)
        temp_dict[str(i) + "," + str(j)] = temp_set.copy()
      
itemset.update(temp_dict)

print(count_items)
print(len(count_items))
```
